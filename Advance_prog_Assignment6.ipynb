```python
def find_the_difference(s,t):
    s_new = list(set(s))
    t_new = list(set(t))
    sorted_s = sorted(s_new)
    sorted_t = sorted(t_new)
    if len(sorted_s) != len(sorted_t):   
        for i,ch in enumerate(sorted_s):
            if ch != sorted_t[i]:
                return ch
        return sorted_t[-1]
    else:
        for i,ch in enumerate(sorted_t):
            if ch != t[i]:
                return ch
        return t[-1]
print(find_the_difference("abcd", "abcde"))
print(find_the_difference("", "y"))
print(find_the_difference("ae", "aea"))
```

```python
def count_datatypes(*args):
    list_order = ["int", "str", "bool", "list", "tuple", "dict"]
    datatype_list = []
    datatype_count_list = []
    for arg in args:
        datatype_list.append(type(arg).__name__)
    for dt in list_order:
        datatype_count_list.append(datatype_list.count(dt))
    return datatype_count_list
print(count_datatypes(1, 45, "Hi", False))
print(count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1))
print(count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23))
print(count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]))
```

```python
def fib_str(inp_num, inp_list):
    first = inp_list[0]
    sec = inp_list[1]
    res = inp_list.copy()
#     print(res)
    for i in range(2,inp_num):
        temp = sec + first
        first = sec
        sec = temp
#         print(temp)
        res.append(temp)
    return ", ".join(res)
print(fib_str(3, ["j", "h"]))
print(fib_str(5, ["e", "a"]))
print(fib_str(6, ["n", "k"]))
```

```python
def fib(n):
    # first two terms
    n1, n2 = 0, 1
    count = 0

    # check if the number of terms is valid
    if n <= 0:
        return ("Fibonacci for negative number is not possible")
    elif n == 1:
        return 1
    else:
        while count < n:
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1
        else:
            return n1
        
print(fib(6))
print(fib(1))
print(fib(2))
print(fib(9))
print(fib(12))
```
